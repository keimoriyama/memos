```python
from matplotlib import pyplot as plt
from random import randint

x = list(range(10))
y1 = [randint(1, 100) for _ in x]
y2 = [randint(1, 100) for _ in x]

fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax.plot(x, y1)
ax = fig.add_subplot(2, 1, 2)
ax.plot(x, y2)
plt.show()
```
